### Plotting specific values

```python
# import the library for plotting
import matplotlib.pyplot as plt 
    
# x axis values 
# this is a "list" of values
x = [1,2,3] 
# corresponding y axis values 
y = [2,4,1] 
    
# plotting the points  
plt.plot(x, y) 
    
# naming the x axis 
plt.xlabel('x - axis') 
# naming the y axis 
plt.ylabel('y - axis') 
    
# giving a title to my graph 
plt.title('My first graph!') 
    
# function to show the plot 
plt.show() 
```

### Defining functions

Now, let's plot a parabola: what does the function look like?

```python
# you can create "functions" that do specific actions for you
# here, the function takes in an input x and returns the value x^2 
def xsquared(x):
    return x**2
```

### Plotting functions

```python
# import the necessary libraries, so the computer knows what to do
import matplotlib.pyplot as plt
import numpy as np

#create some data
# start with your range (x coordinates)
domain = np.linspace(-10,10,100) # this creates a number of points from 1 - 100 spaced by 1
# print(domain)

# for each point, calculate the values for y 
yvals = [xsquared(xvals) for xvals in domain]
plt.plot(domain, yvals)
plt.show()
```
